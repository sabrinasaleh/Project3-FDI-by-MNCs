```python
# Interpolate missing values with method="time" and limit_direction="both"
UK_df = pd.read_csv("Resources/individual_country_missing/1_UnitedKingdom.csv", parse_dates=["year"])
UK_df.set_index("year", inplace=True)
UK_df

# type(UK_df.year[0])
```

```python
UK_df_complete = UK_df.interpolate(method="time", limit_direction="both")
UK_df_complete
# UK_df_complete.to_csv("Resources/individual_country_complete/1_UK_complete.csv")
```
